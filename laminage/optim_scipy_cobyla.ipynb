## Contraintes sur les épaisseurs de sortie

```python
epsilon = 0.01

def constraint_epsort_1(x, epentre_1):
    ct = x[0] - epentre_1*(1-0.52)
    return ct   
def constraint_epsort_1bis(x, epentre_1):
    ct = epentre_1*(1-0.27) - x[0]
    return ct   

def constraint_epsort_2(x):
    ct = x[1] - x[0]*(1-0.48)
    return ct   
def constraint_epsort_2bis(x):
    ct = x[0]*(1-0.2) - x[1]
    return ct   

def constraint_epsort_3(x):
    ct = x[2] - x[1]*(1-0.4)
    return ct   
def constraint_epsort_3bis(x):
    ct = x[1]*(1-0.17) - x[2]
    return ct   


def constraint_epsort_4(x):
    ct = x[3] - x[2]*(1-0.35)
    return ct   
def constraint_epsort_4bis(x):
    ct = x[2]*(1-0.2) - x[3]
    return ct   

def constraint_epsort_5(x):
    ct = x[4] - x[3]*(1-0.32)
    return ct   
def constraint_epsort_5bis(x):
    ct = x[3]*(1-0.18) - x[4]
    return ct   

def constraint_epsort_6(x):
    ct = x[5] - x[4]*(1-0.26)
    return ct   
def constraint_epsort_6bis(x):
    ct = x[4]*(1-0.15) - x[5]
    return ct   

def constraint_epsort_7(x,esort_7):
    ct = esort_7 - x[5]*(1-0.16)
    return ct   
def constraint_epsort_7bis(x,esort_7):
    ct = x[5]*(1-0.06) - esort_7
    return ct   
```

```python

constraints_full = [
    {'type': 'ineq', 'fun': constraint_epsort_1, 'args': (epentr_1,)},
    {'type': 'ineq', 'fun': constraint_epsort_2},
    {'type': 'ineq', 'fun': constraint_epsort_3},
    {'type': 'ineq', 'fun': constraint_epsort_4},
    {'type': 'ineq', 'fun': constraint_epsort_5},
    {'type': 'ineq', 'fun': constraint_epsort_6},
    {'type': 'ineq', 'fun': constraint_epsort_7, 'args': (esort_7,)},
    {'type': 'ineq', 'fun': constraint_epsort_1bis, 'args': (epentr_1,)},
    {'type': 'ineq', 'fun': constraint_epsort_2bis},
    {'type': 'ineq', 'fun': constraint_epsort_3bis},
    {'type': 'ineq', 'fun': constraint_epsort_4bis},
    {'type': 'ineq', 'fun': constraint_epsort_5bis},
    {'type': 'ineq', 'fun': constraint_epsort_6bis},
    {'type': 'ineq', 'fun': constraint_epsort_7bis, 'args': (esort_7,)},
    {'type': 'ineq', 'fun': contrainte_pmax_1, 'args': (modele_f1,epentr_1,larg,metal,pmax['f1'])},
    {'type': 'ineq', 'fun': contrainte_pmax_2, 'args': (modele_f2,larg,metal,pmax['f2'])},
    {'type': 'ineq', 'fun': contrainte_pmax_3, 'args': (modele_f3,larg,metal,pmax['f3'])},
    {'type': 'ineq', 'fun': contrainte_pmax_4, 'args': (modele_f4,larg,metal,pmax['f4'])},
    {'type': 'ineq', 'fun': contrainte_pmax_5, 'args': (modele_f5,larg,metal,pmax['f5'])},
    {'type': 'ineq', 'fun': contrainte_pmax_6, 'args': (modele_f6,larg,metal,pmax['f6'])},
    {'type': 'ineq', 'fun': contrainte_pmax_7, 'args': (modele_f7,esort_7,larg,metal,pmax['f7'])}
]

```

```python
from scipy.optimize import minimize

# Point initial
x0 = [19.5, 14.3,10.1,7.7,5.45,4.35]

# Résolution avec la méthode COBYLA
result = minimize(
    objective_power, 
    x0, 
    args=(modele_ftot, epentr_1,larg,metal,esort_7),  # Passage des paramètres à la fonction objectif
    method='COBYLA',
    constraints=constraints_full
)

# Affichage des résultats
print("Succès :", result.success)
print("Solution trouvée :", result.x)
print("Valeur minimale de la fonction :", result.fun)
```
